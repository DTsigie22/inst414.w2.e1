```python
import pandas as pd
import json
import requests
```

```python
all_art = []
for i in range(5):
    url = "https://api.artic.edu/api/v1/artworks?fields=id,title,artist_display,date_display,place_of_origin,artist_title,artwork_type_title,artist_id,style_id,style_titles&limit=100"
    response = requests.get(url+"&page={}".format(i))
    all_art.append(response.json())
```

```python
art = []
for page in all_art:
    for x in page["data"]:
        art.append(x)
```

```python
len(art)
```

```python
art_list = {}
art_style = {}
art_title = {}
for piece in art:
       art_title[piece['id']] = piece["title"]
for piece in art:
    art_style[piece['id']] = ({})
```

```python
for piece in art:
    if piece["style_titles"] != None:
        for style in piece["style_titles"]:
            art_style[piece['id']][style] = 1
```

```python
index = art_style.keys()
rows = [art_style[k] for k in index]
df = pd.DataFrame(rows, index=index)
df = df.fillna(0)
df
```

```python
from scipy.sparse import lil_matrix
import scipy.spatial.distance
```

```python
target_art_id_1 = 158949
target_art_id_2 = 189294
target_art_id_3 = 152836
```

```python
target_art_style = df.loc[target_art_id_1]
distances = scipy.spatial.distance.cdist(df, [target_art_style], metric = "euclidean") [:,0]
query_distances = list(zip(df.index, distances))
```

```python
for similar_art_id, similar_style in sorted(query_distances, key=lambda x: x[1], reverse=False) [:10]:
    similar_art = art_style[similar_art_id]
    print(similar_art_id, art_title[similar_art_id], similar_style)
```
