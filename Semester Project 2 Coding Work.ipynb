```python
url_standings_2020 = "https://api-nba-v1.p.rapidapi.com/standings"

querystring_20 = {"league":"standard","season":"2020"}

headers_20 = {
	"X-RapidAPI-Key": "dc2bf632a0msh994f55806e2e173p1b0bf9jsn5df0158d073b",
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}

response_stand_2020 = requests.get(url_standings_2020, headers=headers_20, params=querystring_20)

print(response_stand_2020.json())
```

```python
team_names =[]
team_names.append(response_stand_2020.json())

nba_teams = []
for p in team_names:
    for v in p['response']:
        nba_teams.append(v)
print(nba_teams)    
```

```python
df_teams = pd.DataFrame(nba_teams)
df_teams
```

```python
len(df_teams['conference'])
```

```python
team_name_map = {}
for index, row in df_teams.iterrows():
    team_name_map[row['team']['id']] = row['team']['name']
```

```python
team_name_map
```

```python
nba_2020 = []
for nba_teams in all_teams_2020:
    if 'message' in nba_teams.keys():
        continue
    id = nba_teams["parameters"]["id"]
    points = nba_teams['response'][0]['points']
    rebounds = nba_teams['response'][0]['totReb']
    assist = nba_teams['response'][0]['assists']
    fast = nba_teams['response'][0]['fastBreakPoints']
    steal = nba_teams['response'][0]['steals']
    block = nba_teams['response'][0]['blocks']
    nba_2020.append({'id': id, 'Points': points, 'Rebounds': rebounds, 'Assists': assist, 'FastBreak': fast, 'Steals': steal, 'Blocks': block})
df_nba_2020 = pd.DataFrame(nba_2020)
df_nba_2020 = df_nba_2020.set_index("id")
df_nba_2020
```

```python
sil_score_2020 = []
for test_k in range(2, 30, 2):
    print(test_k)
    
    tmp_model_20 = KMeans(n_clusters=test_k)
    tmp_model_20.fit(df_nba_2020)
    
    score_20 = silhouette_score(df_nba_2020, tmp_model_20.labels_)
    sil_score_2020.append((test_k, score_20))
```

```python
sil_1_df = pd.DataFrame(sil_score_2020, columns = ["k", "score"])
sil_1_df
```

```python
k_2020 = 2
cluster_2020 = KMeans(n_clusters = k_2020)
cluster_2020.fit(df_nba_2020)
cluster_label_2020 = cluster_2020.predict(df_nba_2020)
nba2020_cluster_df = pd.DataFrame(cluster_label_2020, index= df_nba_2020.index, columns = ["clusters"])
```
