```python
import requests
import pandas as pd
import csv
```

```python
url = "https://www.freetogame.com/api/games"
response = requests.get(url)
print(response)
```

```python
games = response.json()
games
```

```python
genre_map = {}
publisher_map = {}
date_map = {}
game_name_map = {}


for lines in games:
        game_name_map[lines["id"]] = lines["title"]
            
        # For each actor, add this movie's genres to that actor's list
        this_games_genres = genre_map.get(lines["id"], {})
            
            # Increment the count of genres for this actor
        this_games_genres[lines["genre"]] = this_games_genres.get(lines["genre"], 0) + 1
        this_games_genres[lines["publisher"]] = this_games_genres.get(lines["publisher"], 0) + 1
        
            # Update the map
        genre_map[lines["id"]] = this_games_genres
```

```python
genre_map[23]
```

```python
df = pd.DataFrame(genre_map.values(), index = genre_map.keys())
df= df.fillna(0)
df
```

```python
from sklearn.cluster import KMeans
```

```python
k = 8
```

```python
cluster = KMeans(n_clusters = k)
cluster.fit(df)
```

```python
cluster_label = cluster.predict(df)
games_cluster_df = pd.DataFrame(cluster_label, index=df.index, columns = ["clusters"])
```

```python
games_cluster_df
```

```python
games_cluster_df["clusters"].value_counts()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(df, cluster.labels_)
```

```python
silh_scores = []
for test_k in range(2, 30, 2):
    print(test_k)
    
    model = KMeans(n_clusters=test_k)
    model.fit(df)
    
    score = silhouette_score(df, model.labels_)
    silh_scores.append((test_k, score))
```

```python
silh_df = pd.DataFrame(silh_scores, columns=["k", "score"])
silh_df
```

```python
for clusters,game in games_cluster_df.groupby("clusters"):
    print("Cluster:", clusters, "Size:", game.shape[0])
    
    for game_id in game.sample(10).index:
        print("\t", game_id, game_name_map[game_id])
```
